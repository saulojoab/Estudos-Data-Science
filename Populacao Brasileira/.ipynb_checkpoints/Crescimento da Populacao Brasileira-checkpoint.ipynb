# ESTUDO DE CASO
### Crescimento da População Brasileira
de 1980 - 2016 (DataSUS)


```python
#coding: utf-8
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import matplotlib.pyplot as plt

# Abrindo o dataset da população brasileira.
dados = open("populacao-brasileira.csv").readlines();

x = [];
y = [];

# Percorrendo as linhas.
for i in range(len(dados)):
    # Ignorando a primeira linha.
    if (i != 0):
        linha = dados[i].split(';');
        # Ano.
        x.append(int(linha[0]));
        # População.
        y.append(int(linha[1]));

# GRÁFICO DE LINHA.
# Título.
plt.title("Crescimento da População Brasileira - 1980 até 2016");
        
# Labels.
plt.xlabel("Ano");
plt.ylabel("População x100.000.000");
        
# Plotando.
plt.plot(x, y);
plt.show();
    
```

```python
# GRÁFICO DE BARRAS.
# Título.
plt.title("Crescimento da População Brasileira - 1980 até 2016");
        
# Labels.
plt.xlabel("Ano");
plt.ylabel("População x100.000.000");
        
# Plotando.
plt.bar(x, y);
plt.show();
```

```python
# COMBINANDO GRÁFICO DE LINHAS E DE BARRAS.
# Título.
plt.title("Crescimento da População Brasileira - 1980 até 2016");
        
# Labels.
plt.xlabel("Ano");
plt.ylabel("População x100.000.000");
        
# Plotando a combinação de grafos.
plt.plot(x, y, linestyle = "--");
plt.bar(x, y, color = "#e4e4e4");
plt.show();
```
