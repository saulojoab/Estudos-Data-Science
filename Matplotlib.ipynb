```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import matplotlib.pyplot as plt
import numpy as np
```

```python
# ESTUDANDO MATPLOTLIB
# GR√ÅFICO DE LINHA
x = [1, 2, 5]
y = [2, 3, 7]

# T√≠tulo.
plt.title("Meu primeiro gr√°fico!");

# Eixos.
plt.xlabel("Eixo X");
plt.ylabel("Eixo Y");

# Precisa de x e y.
plt.plot(x, y);

# Mostrando.
plt.show();
```

```python
# GR√ÅFICO DE BARRAS
x = [1, 2, 3, 4, 5];
y = [3, 6, 2, 1, 4];

# Legendas.
plt.title("Gr√°fico de Barras");

plt.xlabel("Eixo X");
plt.ylabel("Eixo Y");

# Plotando o gr√°fico de barras.
plt.bar(x, y);
plt.show();
```

```python
# GR√ÅFICO DE BARRAS COMPARATIVO.
# √çmpares.
x1 = [1, 3, 5, 7, 9];
y1 = [3, 6, 2, 1, 4];

# Pares.
x2 = [2, 4, 6, 8, 10];
y2 = [5, 1, 3, 7, 4];

# Legendas.
plt.title("Gr√°fico de Barras");

plt.xlabel("Eixo X");
plt.ylabel("Eixo Y");

# Plotando o gr√°fico de barras.
plt.bar(x, y, label = "Grupo 1");
plt.bar(x2, y2, label = "Grupo 2");

# Mostrando as labels.
plt.legend();

plt.show();
```

```python
# SCATTERPLOT (GR√ÅFICOS DE PONTOS)

x = [1, 3, 5, 7, 9];
y = [3, 6, 2, 1, 4];

# Legendas.
plt.title("Gr√°fico de Pontos");

plt.xlabel("Eixo X");
plt.ylabel("Eixo Y");

# Plotando o gr√°fico de pontos.
# Podemos customizar os atributos: cor, tamanho, formato, tipo de linha e etc. 
plt.scatter(x, y, label = "Meus Pontos", color = "R");

# Plotando linhas que ligam os pontos.
plt.plot(x, y);

# Mostrando da label.
plt.legend();

# Mostrando o gr√°fico.
plt.show();
```
