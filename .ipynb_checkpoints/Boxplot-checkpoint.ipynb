# Boxplot - Diagrama de Caixa
Serve pra visualizar quão dispersos estão os dados.

<b>Maximum</b>: Valor máximo.<br>
<b>IRQ</b>: Contêm 50% dos valores.<br>
<b>Minimum</b>: Valor mínimo.<br>
<b>Outline</b>: Valores muito dispersos.<br>




```python
#coding: utf-8
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import matplotlib.pyplot as plt
import random

# Criando um vetor vazio.
vetor = [];

# Adicionando valores no vetor.
for i in range(10):
    vetor.append(random.randint(0, 10000));

# BOXPLOT
# Título.
plt.title("Boxplot");

# Plotando.
plt.boxplot(vetor);
plt.show();

```
